something to pull the root element structure (rdf:RDF or thredds:catalog or just catalog), namespaces (used or not), urls

```python
from lxml import etree
import os
import json
import glob

files = glob.glob('../testdata/solr_20150320/clean_20150325/*.json')

parser = etree.XMLParser(
            encoding='utf-8',
            remove_blank_text=True,
            remove_comments=True,
            recover=True
        )

def get_namespaces(xml):
    if xml is None:
        return {}
    
    document_namespaces = dict(xml.xpath('/*/namespace::*'))
    if None in document_namespaces:
        document_namespaces['default'] = document_namespaces[None]
        del document_namespaces[None]

    # now run through any child namespace issues
    all_namespaces = xml.xpath('//namespace::*')
    for i, ns in enumerate(all_namespaces):
        if ns[1] in document_namespaces.values():
            continue
        new_key = ns[0] if ns[0] else 'default%s' % i
        document_namespaces[new_key] = ns[1]
    
    return document_namespaces

def prefix(tag, namespaces):
    for prefix, ns in namespaces.iteritems():
        wrapped_ns = '{%s}' % ns
        tag = tag.replace(wrapped_ns, prefix + ':')
    if '{' in tag and '}' in tag:
        return ''

    return tag
    
with open('../testdata/solr_20150320/identity_inputs.csv', 'w') as f:
    f.write('url,root,prefixes,uris\n')
    
for f in files:
    with open(f, 'r') as g:
        data = json.loads(g.read())
    
    content = data['content'].encode('unicode_escape')
    url = data['source_url']
    digest = data['digest']
    
    xml = etree.fromstring(content, parser=parser)
    if xml is None:
        continue
        
    namespaces = get_namespaces(xml)
    
    root = xml.xpath('/*')[0].tag
    prefixed_root = prefix(root, namespaces)
    
    # print url, prefixed_root, namespaces
    with open('../testdata/solr_20150320/identity_inputs.csv', 'a') as f:
        f.write(','.join([digest,
                          url, 
                          prefixed_root, 
                          '|'.join(namespaces.keys()),
                          '|'.join(namespaces.values())
                          ]) + '\n')
    
```
