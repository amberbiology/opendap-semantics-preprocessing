```python
import glob
import json
import urlparse

files = glob.glob('../testdata/solr_20150320/identify_20150325_p/*.json')

def parse_url(url):
    parsed = urlparse.urlparse(url)
    
    base_url = urlparse.urlunparse((
            parsed.scheme,
            parsed.netloc,
            parsed.path,
            parsed.params,
            None,
            None
        ))
    
    # return base_url, urlparse.parse_qs(parsed.query)
    return base_url, parsed.query

services = {}
protocols = {}
for f in files[:100]:
    with open(f, 'r') as g:
        data = json.loads(g.read())
    
    url = data['source_url']
    identity = data['identity']
    
    parsed_url, params = parse_url(url)
    
    # let's just ignore anything not identified (we got plenty)
    if not identity['protocol']:
        continue
    
#     # by base_url
#     profiles = services.get(parsed_url, [])
#     profiles.append({"parameters": params, 
#                      "protocol": identity['protocol'], 
#                      "service": identity['service'],
#                      "version": identity['version']
#                     })
#     services[parsed_url] = profiles
    
    # and let's do one (for prettiness and collapsing)
    # by protocol
    protocol = identity['protocol']
    service = identity['service']
    service = service if service else 'ALL'
    
    protocol_dict = protocols.get(protocol, {})
    service_dict = protocol_dict.get(service, {})
    routes = service_dict.get(parsed_url, [])
    if params:
        routes.append(params)
    service_dict[parsed_url] = routes
    protocol_dict[service] = service_dict
    protocols[protocol] = protocol_dict
    

# print [{s: v} for s, v in services.iteritems()]
print json.dumps(protocols, indent=2)
```

```python
# and remap protocols to the name+children widget in d3
# in the worst possible way, don't care. 

clusters = []

for k, v in protocols.iteritems():
    children = []
    for sk, sv in v.iteritems():
        other_children = []
        for rk, rv in sv.iteritems():
             other_children.append({"name": rk, "children": rv})
        
        children.append({"name": sk, "children": other_children})
    
    clusters.append({"name": k, "children": children})
    
clusters = {"name": "identities", "children": clusters}

# print json.dumps(clusters, indent=2)
```

```python
with open('protocol_tree.json', 'w')as g:
    g.write(json.dumps(clusters, indent=2))
```

```python
import urlparse

def jaccardify(a, b):
    # except the lengths MUST be the same so append empty things to
    # the shorted thing (for scikit anyway)
    if len(a) != len(b):
        to_append = [''*abs(len(a) - len(b))]
    
    if len(a) < len(b):
        a += to_append
    if len(b) < len(a):
        b += to_append
    
    return (len(set(a) & set(b))) / float(len(set(a) | set(b)))

def compare_things(url, potentials):
    jaccards = []
    for potential in potentials:
        url_path = urlparse.urlparse(url).path.split('/')
        potential_path = urlparse.urlparse(potential).path.split('/')
        j = jaccardify(url_path, potential_path)
        jaccards.append((potential, j))
    
    return {"jaccards": jaccards}

def return_thing(s, arr):
    # yeah return the thing from a list of things
    return next(
        iter(
            a for a in arr if a['path'] == s
        ), {}
    )
    

# a not good list from opendap and a random gbif thing
acdiscs = [
'http://acdisc.sci.gsfc.nasa.gov/opendap/ncml/Aqua_AIRS_Level3/AIRX3STM.006/2015/AIRS.2015.02.01.L3.RetStd028.v6.0.11.0.G15076164741.hdf.ncml',
'http://acdisc.sci.gsfc.nasa.gov/opendap/ncml/Aqua_AIRS_Level3/AIRX3STM.006/2015/AIRS.2015.02.01.L3.RetStd028.v6.0.11.0.G15076164741.hdf.ncml.ddx',
'http://acdisc.sci.gsfc.nasa.gov/opendap/ncml/Aqua_AIRS_Level3/AIRX3STM.006/2015/catalog.xml',
'http://acdisc.sci.gsfc.nasa.gov/opendap/ncml/Aqua_AIRS_Level3/AIRX3STM.006/catalog.xml',
'http://acdisc.sci.gsfc.nasa.gov/opendap/Nimbus7_TOMS_Level3/TOMSN7L3.008/catalog.xml',
'http://api.gbif.org/v1/dataset/50c9509d-22c7-4a22-a47d-8c48425ef4a7/document'
]

buckets = {}

# so bucket by host with a dict per path that contains "things i am similar to"

for ac in acdiscs:
    parsed = urlparse.urlparse(ac)
    
    host = parsed.netloc
    paths = parsed.path.split('/')
    
    knowns = buckets.get(host, [])
    
    thing = return_thing(parsed.path, knowns)
    if not thing:
        thing = {"path": parsed.path}
        
    
    comparisons = compare_things(ac, acdiscs)
    thing.update({"comparisons": comparisons})

    knowns.append(thing)
    buckets[host] = knowns

buckets    
```

```python
import json
import glob
import os
from simhash import Simhash, SimhashIndex

files = glob.glob('../testdata/solr_20150320/identify_20150325_p/*.json')

def parse_url(url):
    parsed = urlparse.urlparse(url)
    
    base_url = urlparse.urlunparse((
            parsed.scheme,
            parsed.netloc,
            parsed.path,
            parsed.params,
            None,
            None
        ))
    
    # return base_url, urlparse.parse_qs(parsed.query)
    return base_url, parsed.query

urls = []
for f in files:
    with open(f, 'r') as g:
        data = json.loads(g.read())
    
    url = data['source_url']
    identity = data['identity']
    
    # let's just ignore anything not identified (we got plenty)
    if not identity['protocol']:
        continue
 
    protocol = identity['protocol']
    
    urls.append((url, protocol))
    

print len(urls)    
    
```

```python
import time

# let's generate the simhash singletons
# and generate the index (this cannot be performant)
test_index = [(u[0], Simhash(u[0])) for u in urls]

index = SimhashIndex(test_index, k=20)

start_time = time.time()
duplicates = []
for test_url, test_simhash in test_index:
    dupes = index.get_near_dups(test_simhash)
    
    if len(dupes) > 0:
        duplicates.append((test_url, dupes))

print 'takes:', time.time() - start_time
print len(duplicates)     
print 
duplicates[:5]
```
