#### round one of relationship building

Let's start to sort out what we have based on base URLs (no query parameters) and profiles (harvested, not generated from the response).


```python
import glob
import json
import urlparse

files = glob.glob('../testdata/solr_20150320/identify_20150325_p/*.json')

def parse_url(url):
    parsed = urlparse.urlparse(url)
    
    base_url = urlparse.urlunparse((
            parsed.scheme,
            parsed.netloc,
            parsed.path,
            parsed.params,
            None,
            None
        ))
    
    # return base_url, urlparse.parse_qs(parsed.query)
    return base_url, parsed.query

services = {}
protocols = {}
for f in files[:100]:
    with open(f, 'r') as g:
        data = json.loads(g.read())
    
    url = data['source_url']
    identity = data['identity']
    
    parsed_url, params = parse_url(url)
    
    # let's just ignore anything not identified (we got plenty)
    if not identity['protocol']:
        continue
    
#     # by base_url
#     profiles = services.get(parsed_url, [])
#     profiles.append({"parameters": params, 
#                      "protocol": identity['protocol'], 
#                      "service": identity['service'],
#                      "version": identity['version']
#                     })
#     services[parsed_url] = profiles
    
    # and let's do one (for prettiness and collapsing)
    # by protocol
    protocol = identity['protocol']
    service = identity['service']
    service = service if service else 'ALL'
    
    protocol_dict = protocols.get(protocol, {})
    service_dict = protocol_dict.get(service, {})
    routes = service_dict.get(parsed_url, [])
    if params:
        routes.append(params)
    service_dict[parsed_url] = routes
    protocol_dict[service] = service_dict
    protocols[protocol] = protocol_dict
    

# print [{s: v} for s, v in services.iteritems()]
print json.dumps(protocols, indent=2)
```

```python
# and remap protocols to the name+children widget in d3
# in the worst possible way, don't care. 

clusters = []

for k, v in protocols.iteritems():
    children = []
    for sk, sv in v.iteritems():
        other_children = []
        for rk, rv in sv.iteritems():
             other_children.append({"name": rk, "children": rv})
        
        children.append({"name": sk, "children": other_children})
    
    clusters.append({"name": k, "children": children})
    
clusters = {"name": "identities", "children": clusters}

# print json.dumps(clusters, indent=2)
```

```python
with open('protocol_tree.json', 'w')as g:
    g.write(json.dumps(clusters, indent=2))
```
